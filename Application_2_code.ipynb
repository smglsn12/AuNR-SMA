```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import joblib
import sklearn as sk
```

```python
db1 = pd.read_csv('Nimbus_070721_for_jakob.csv')
db1
```

```python
db2  = joblib.load('Spectra_amts_predicted_sizes.joblib')
db2
```

```python
db2n = db2.replace('NA',np.nan).dropna(axis = 0)
db2n
```

```python
all_db = pd.concat([db1,db2n])
all_db = all_db.dropna(axis = 1)
all_db
```

```python
all_db = all_db.sample(frac = 1.0)
all_db
```

```python
inputs = all_db.loc[:,['H2O','CTAB','Au','Ag','HQ','NaBH4','NaBH4_2nd']]
outputs = all_db.loc[:,['Predicted Length Mean','Predicted Length Std','Predicted Diameter Mean','Predicted Diameter Std','Predicted AR Mean','Predicted AR Std']]
```

```python
from sklearn.model_selection import GridSearchCV as gscv
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet as ENET
from sklearn.linear_model import BayesianRidge as BR
from sklearn.kernel_ridge import KernelRidge as KRR
from sklearn.gaussian_process import GaussianProcessRegressor as GPR
from sklearn.neighbors import KNeighborsRegressor as KNR
from sklearn.svm import SVR as SVR
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.ensemble import GradientBoostingRegressor as GBR
```

```python
param_dic = {
    'ENET' : {
        'alpha' : [0.0001,0.001,0.01,0.1,1.0,10.0],
        'l1_ratio' : [0.0,0.1,0.25,0.5,0.75,0.9,1.0]  
    },
    'BR' : {
        'alpha_init' : [1.0e-9,1.0e-4,1.0,1.0e3],
        'lambda_init' : [1.0e-9,1.0e-4,1.0,1.0e3]
    },
    'KRR' : { 
        'kernel' : ['linear','rbf','sigmoid'],
        'alpha' : [0.001,0.1,1.0],
        'gamma' : [1.0e-6,1.0e-3,1.0,1.0e3]
        
    },
    'GPR' : {
        'alpha' : [1.0e-12,1.0e-9,1.0e-6,1.0e-3,1.0]   
    },
    'KNR' : {
        'n_neighbors' : [3,5,8,13],
        'p' : [1.5,2.0,3.0,5.0]
        
    },
    'SVR' : {
        'kernel' : ['linear','rbf','sigmoid'],
        'epsilon' : [0.1,1.0],
        'C' : [1.0e-6,1.0e-3,1.0,1.0e3],
        'gamma' : [1.0e-6,1.0e-3,1.0]
    },
    'RFR' : {
        'n_estimators' : [30,100,200,500],
        'min_samples_leaf' : [1,2,3,5],
        'max_features' : [0.3,0.6,1.0]
    },
    'GBR' : {
        'alpha' : [0.01,0.1,0.5,0.9],
        'n_estimators' : [30,100,200,500],
        'min_samples_leaf' : [1,2,3,5],
        'learning_rate' : [0.01,0.03,0.1,0.3]
    },
}
```

```python
def fit_all_models(output_vector,inputs = inputs, param_dic = param_dic):
    all_results = []
    for model in param_dic.keys():
        if model == 'ENET':
            model_actual = ENET()
        elif model == 'BR':
            model_actual = BR()
        elif model == 'KRR':
            model_actual = KRR()
        elif model == 'GPR':
            model_actual = GPR()
        elif model == 'KNR':
            model_actual = KNR()
        elif model == 'SVR':
            model_actual = SVR()    
        elif model == 'RFR':
            model_actual = RFR()
        elif model == 'GBR':
            model_actual = GBR()
        gs_model = gscv(model_actual,param_grid = param_dic[model],scoring = 'neg_mean_absolute_percentage_error')
        gs_model.fit(inputs,output_vector)
        all_results.append([gs_model,gs_model.cv_results_])
    return all_results
        
```

```python
gbr = GBR()
gbr.fit(inputs,outputs.iloc[:,5])
r2_score(gbr.predict(inputs),outputs.iloc[:,5])
gs_model = gscv(gbr,param_grid = param_dic['GBR'],scoring = 'r2')
gs_model.fit(inputs,outputs.iloc[:,0])
```

```python
print(gs_model.best_estimator_.get_params())
print(gs_model.best_score_)
```

```python
all_out_all_models = joblib.Parallel(n_jobs = len(outputs.columns)) (joblib.delayed(fit_all_models)(outputs[cx]) for cx in outputs.columns)
```

```python
def fit_all_models_mse(output_vector,inputs = true_inputs, param_dic = param_dic):
    all_results = []
    for model in param_dic.keys():
        if model == 'ENET':
            model_actual = ENET()
        elif model == 'BR':
            model_actual = BR()
        elif model == 'KRR':
            model_actual = KRR()
        elif model == 'GPR':
            model_actual = GPR()
        elif model == 'KNR':
            model_actual = KNR()
        elif model == 'SVR':
            model_actual = SVR()    
        elif model == 'RFR':
            model_actual = RFR()
        elif model == 'GBR':
            model_actual = GBR()
        gs_model = gscv(model_actual,param_grid = param_dic[model],scoring = 'neg_mean_squared_error')
        gs_model.fit(inputs,output_vector)
        all_results.append([gs_model,gs_model.cv_results_])
    return all_results
        
```

```python
all_out_all_models_tipmse = joblib.Parallel(n_jobs = len(outputs.columns)) (joblib.delayed(fit_all_models_mse)(outputs[cx]) for cx in outputs.columns)
```

```python
all_resultstipmse = []
for out in all_out_all_models_tipmse:
    results = []
    for model in out:
        results.append(np.sqrt(-model[0].best_score_))
    all_resultstipmse.append(results)
print(all_resultstipmse)      
```
