```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import joblib
import sklearn as sk

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['font.sans-serif'] = "Arial"
mpl.rcParams['font.family'] = "sans-serif"
mpl.rcParams['font.size'] = 16
```

```python
db1 = pd.read_csv('Nimbus_070721_for_jakob.csv')
db1
```

```python
db2  = joblib.load('Spectra_amts_predicted_sizes.joblib')
db2
```

```python
db2n = db2.replace('NA',np.nan).dropna(axis = 0)
db2n
```

```python
all_db = pd.concat([db1,db2n])
all_db = all_db.dropna(axis = 1)
all_db
```

```python
import random
```

```python
random.seed(17)
```

```python
all_db = all_db.sample(frac = 1.0)
all_db
```

```python
inputs = ['H2O','CTAB','Au','Ag','HQ','NaBH4','NaBH4_2nd']
outputs = ['Predicted Length Mean','Predicted Length Std','Predicted Diameter Mean','Predicted Diameter Std','Predicted AR Mean','Predicted AR Std']
```

```python
from sklearn.model_selection import GridSearchCV as gscv
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.model_selection import train_test_split as tts
```

```python
RFR_params = {
        'n_estimators' : [100],
        'min_samples_leaf' : [1,2,3,5,8],
        'max_features' : [0.1,0.3,0.6,1.0] }
```

```python
results = []

for out in outputs:
    X_train,X_test,y_train,y_test = tts(all_db[inputs],all_db[out],random_state = 34)
    model = gscv(RFR(),param_grid = RFR_params,scoring = 'neg_mean_squared_error')
    model.fit(X_train,y_train)
    results.append([model,y_test,y_train,model.best_estimator_.predict(X_test),model.best_estimator_.predict(X_train)])
    
```

```python
for result in results:
    plt.title(outputs[results.index(result)])
    plt.scatter(result[2],result[4],label = 'training')
    plt.scatter(result[1],result[3],label = 'testing')
    plt.xlabel('Calculated Values')
    plt.ylabel('Predicted Values')
    max_value = max([max(n) for n in result[1:]])
    min_value = min([min(n) for n in result[1:]])
    plt.plot(np.linspace(min_value,max_value,10),1.2*np.linspace(min_value,max_value,10),color = 'k',linestyle = '--',label = '20% error')
    plt.legend()

    plt.plot(np.linspace(min_value,max_value,10),np.linspace(min_value,max_value,10),color = 'k')
    
    plt.plot(np.linspace(min_value,max_value,10),0.8*np.linspace(min_value,max_value,10),color = 'k',linestyle = '--')
    plt.xlim(min_value,max_value)
    plt.ylim(min_value,max_value)
    
    plt.savefig('ML_' + str(outputs[results.index(result)]) + '.pdf', format = 'pdf')
    plt.show()
```

```python
results_lim = []

for out in outputs:
    X_train,X_test,y_train,y_test = tts(all_db[['CTAB','Ag','HQ','NaBH4','NaBH4_2nd']],all_db[out],random_state = 34)
    model = gscv(RFR(),param_grid = RFR_params,scoring = 'neg_mean_squared_error')
    model.fit(X_train,y_train)
    results_lim.append([model,y_test,y_train,model.best_estimator_.predict(X_test),model.best_estimator_.predict(X_train)])
    
```

```python
for result in results_lim:
    plt.title(outputs[results_lim.index(result)])
    plt.scatter(result[2],result[4],label = 'training')
    plt.scatter(result[1],result[3],label = 'testing')
    plt.xlabel('Calculated Values')
    plt.ylabel('Predicted Values')
    max_value = max([max(n) for n in result[1:]])
    min_value = min([min(n) for n in result[1:]])
    plt.plot(np.linspace(min_value,max_value,10),1.2*np.linspace(min_value,max_value,10),color = 'k',linestyle = '--',label = '20% error')
    plt.legend()

    plt.plot(np.linspace(min_value,max_value,10),np.linspace(min_value,max_value,10),color = 'k')
    
    plt.plot(np.linspace(min_value,max_value,10),0.8*np.linspace(min_value,max_value,10),color = 'k',linestyle = '--')
    plt.xlim(min_value,max_value)
    plt.ylim(min_value,max_value)
    
    plt.savefig('ML_lim_' + str(outputs[results_lim.index(result)]) + '.pdf', format = 'pdf')
    plt.show()
```
